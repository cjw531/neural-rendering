```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/home/jiwon/workspace/img_dataset/train_cropped/images/Registered0000.png")
output = image.copy()
cv2.circle(output, (98, 125),
        radius=2, color=(255, 0, 0), thickness=2) # left top
cv2.circle(output, (98, 985),
        radius=2, color=(255, 0, 0), thickness=2) # left bottom
cv2.circle(output, (1040, 125),
        radius=2, color=(255, 0, 0), thickness=2) # top right
cv2.circle(output, (1040, 985),
        radius=2, color=(255, 0, 0), thickness=2) # top bottom
cv2.imwrite("./example/registered_corners.png", output)
plt.imshow(output)
plt.show()
```

```python
'''
Using corner-marked images will display the 4 corners, but will be removed in the actual preprocessing.
'''
image = cv2.imread("./example/registered_corners.png")
output = image.copy()
# img crop reference: https://stackoverflow.com/a/54885575
output = output[125 : 985 + 1, 98 : 1040 + 1]
cv2.imwrite("./example/registered_cropped.png", output)
plt.imshow(output)
plt.show()
```
